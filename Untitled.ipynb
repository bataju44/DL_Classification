```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

sig_file = '/home/bataju/dlanalysis/datasets-csv/341524.csv'
bkg_file = '/home/bataju/dlanalysis/datasets-csv/410000.csv'
```

```python
_sig = pd.read_csv(sig_file)
_bkg = pd.read_csv(bkg_file)
```

```python
df_sig = _sig.assign(s_b = lambda x: 1)
```

```python
df_bkg= _bkg.assign(s_b = lambda x: 0)
```

```python
result = [df_sig,df_bkg]
df = pd.concat(result)
```

```python
var_list = ["event_number","weight_mc","dphi_MET_tau","tau_0_pt","met_et","bjet_0_pt","dR_tau_bjet","dphi_MET_bjet","upsilon","s_b"]
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df
```

```python
df_array = np.array(df,dtype='float64')
df_feature= df_array[:,2:9]
df_signal = df_array[:,-1:]
df_feature.shape
```

```python
from sklearn import preprocessing


df_norm = preprocessing.normalize(df_feature,axis=0,norm='max')
```

```python
print 'max =  ', np.amax(df_norm,axis=0)
print " "
print 'min =  ',np.amin(df_norm,axis=0)
print " "
print 'mean= ',np.mean(df_norm,axis=0)
print " "
print 'std = ',np.std(df_norm,axis=0)
print " "

```

```python
N_Max =863466 
N_Train = 690773

Train_Sample = df_norm[:N_Train]
Test_Sample = df_norm[N_Train:]

X_Train = Train_Sample
y_Train = df_signal[:N_Train]

X_Test = Test_Sample
y_Test = df_signal[N_Train:]

X_Train.shape
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation,Dropout
from keras.optimizers import SGD


model = Sequential()
model.add(Dense(64, input_dim=7, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

model.fit(X_Train, y_Train,
          epochs=20,
          batch_size=128)
score = model.evaluate(X_Test, y_Test, batch_size=128)
```

```python
from sklearn.metrics import roc_curve

y_pred = model.predict(X_Test).ravel()
fpr_,tpr_, threshold = roc_curve(y_Test,y_pred)
```

```python
from sklearn.metrics import auc
auc_ = auc(fpr_,tpr_)
auc_
```

```python
plt.figure()
plt.plot(fpr_,tpr_,label='Area {:.3f}'.format(auc_))
plt.legend()
plt.xlabel("Fasle positive rate")
plt.ylabel("True positive rate")
plt.show()
```
